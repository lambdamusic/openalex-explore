## Reload JSON data from local

```python
# to load
df_topics = pd.read_json("openalex_topics_2024_09.json")
df_fields = pd.read_json("openalex_fields_2024_09.json")
df_subfields = pd.read_json("openalex_subfields_2024_09.json")
df_domains = pd.read_json("openalex_domains_2024_09.json")
```

## Turn into a foamtree structure / take 1

```python
# df = df.reset_index()  # make sure indexes pair with number of rows
# {'id': ... 'display_name': ... children: []}

df = df_topics
out = []
level_name = "subfield" # "subfield"

for index, row in df.head(50000).iterrows():
    child = {"id" : row['id'], "label" : row['display_name'], "weight": row["works_count"]}
    

    parent_id = row[level_name]['id']

    flag = False
    for el in out:
        if el['id'] == parent_id:
            el["groups"].append(child)
            flag = True
            break

    if not flag: 
        new_parent = {"id" : parent_id, "label" : row[level_name]['display_name']}
        new_parent["groups"] = [child]
        # print("new_parent", parent_id)
        out.append(new_parent)
        


# validation
print("Top level items: ", len(out))
```

## Concat the dataframes

```python
# simply concat all dataframes for easier processings

# add a column 
df_domains['entity_type'] = "domains"
df_fields['entity_type'] = "fields"
df_subfields['entity_type'] = "subfields"
df_topics['entity_type'] = "topics"

df_merged = pd.concat([df_domains, df_fields, df_subfields, df_topics], ignore_index=True)

df_merged.to_json("openalex_merged_cats_2024_09.json")

df_merged.head()
```
