# Topics and other classifications in OpenAlex

Notebook pulling all topics data and turning it into a `js` tree structure that can be used by [FoamTree](https://get.carrotsearch.com/foamtree/latest/demos/). 

```python
import pandas as pd

# !pip install pyalex
from pyalex import Works, Authors, Sources, Institutions, Topics, Publishers, Funders, Fields, Domains, Subfields
import pyalex

# pyalex.config.email = "mail@example.com"
from pyalex import config

config.max_retries = 0
config.retry_backoff_factor = 0.1
config.retry_http_codes = [429, 500, 503]
```

## Reload JSON data from local

```python
# load previously saved files
df_topics = pd.read_json("data/openalex_topics_2024_09.json")
df_fields = pd.read_json("data/openalex_fields_2024_09.json")
df_subfields = pd.read_json("data/openalex_subfields_2024_09.json")
df_domains = pd.read_json("data/openalex_domains_2024_09.json")
```

## Concat the dataframes

Simply concat all dataframes containing topics, domains etc.. for easier processings. 

```python

# add a 'type' column 
df_domains['entity_type'] = "domains"
df_fields['entity_type'] = "fields"
df_subfields['entity_type'] = "subfields"
df_topics['entity_type'] = "topics"

df_merged = pd.concat([df_domains, df_fields, df_subfields, df_topics], ignore_index=True)

df_merged.to_json("data/openalex_merged_cats_2024_09.json")

df_merged.head()
```

## Turn into a foam tree data structure

```python
# structure roughly looks like this
# {'id': ... , 'label': ...  , "weight": .... ,  groups: []}
# see https://get.carrotsearch.com/foamtree/latest/demos/quick-start.html for an example
#
# tip - use the df.head() control for testing with less data
#
# foamtree json shape
out = {"id":"1",
       "label":"Open Alex Topics",
       "groups": [] }

# flag to allow multiple runs of this cell to generate different versions of the dataset  
# topics have 'keywords' as a flat list, we can add them to the tree as a leaf level
INCLUDE_KEYWORDS = False

# domains
df = df_merged[df_merged.entity_type == 'domains']
for index, row in df.head(50).iterrows():
    item = {"id" : row['id'], "label" : row['display_name'], "desc" : row['description'], "weight": row["works_count"], "groups": [] }
    out['groups'].append(item)


# fields
df = df_merged[df_merged.entity_type == 'fields']
for index, row in df.head(500).iterrows():
    item = {"id" : row['id'], "label" : row['display_name'], "desc" : row['description'], "weight": row["works_count"], "groups": []}

    # add to parent
    parent_id = row["domain"]['id']
    for domain in out['groups']:
        if domain['id'] == parent_id:
            domain["groups"].append(item)

# subfields
df = df_merged[df_merged.entity_type == 'subfields']
for index, row in df.head(500).iterrows():
    item = {"id" : row['id'], "label" : row['display_name'], "desc" : row['description'], "weight": row["works_count"], "groups": []}

    # add to parent
    parent_id = row["field"]['id']
    for domain in out['groups']:
        for field in domain['groups']:
            if field['id'] == parent_id:
                field["groups"].append(item)

# topics
df = df_merged[df_merged.entity_type == 'topics']
for index, row in df.head(50000).iterrows():
    item = {"id" : row['id'], "label" : row['display_name'],"desc" : row['description'],  "weight": row["works_count"], "groups": []}

    if INCLUDE_KEYWORDS:
        # keywords_l = []
        for k in row['keywords']:
            # eg "https://openalex.org/keywords/cardiac-imaging"
            k_id = f"""https://openalex.org/keywords/{k.replace(" ", "-").lower()}"""
            item["groups"].append({"id" : k_id, "label" : k, "desc" : f"{k} keyword (no description)" ,  "weight": 0})
        

    # add to parent
    parent_id = row["subfield"]['id']
    for domain in out['groups']:
        for field in domain['groups']:
            for subfield in field['groups']:
                if subfield['id'] == parent_id:
                    subfield["groups"].append(item)





# validation
print("Top level items: ", len(out['groups']))
```

### Export to a JS file for FoamTree

```python
import json

template = """
modelDataAvailable({data_json})
"""
contents = template.format(data_json = json.dumps(out))

if INCLUDE_KEYWORDS:
    with open('data/openalex-topics-keywords-foamtree.js', 'w') as f:
        f.write(contents)
else:
    with open('data/openalex-topics-foamtree.js', 'w') as f:
        f.write(contents)
```
